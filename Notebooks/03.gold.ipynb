```python
import pandas as pd
```

```python
def leitura_csv(file_name):
    return pd.read_csv(f"..\\CSV\\02.silver\\silver_{file_name}.csv")

def salvamento_csv(dataframe, file_name):
    dataframe.to_csv(f"..\\CSV\\03.gold\\gold_{file_name}.csv", index=False)
```

```python
df_individual = leitura_csv("desempenho_individual")
df_times = leitura_csv("desempenho_times")
df_info = leitura_csv("info_partidas")
```

```python
df_sumula_completa = pd.merge(df_individual, df_times, on=["Baba", "Time"], how="inner")
df_sumula_completa
```

```python
df_parcerias = (pd.merge(df_sumula_completa[["Baba", "Jogador", "Time", "Campeao", "Vitorias", "Empates", "Derrotas"]], 
                         df_individual[["Baba", "Jogador", "Time"]], 
                         on=["Baba", "Time"], 
                         how="inner"))
df_parcerias = df_parcerias[df_parcerias["Jogador_x"] != df_parcerias["Jogador_y"]]
df_parcerias = df_parcerias.rename(columns={"Jogador_x": "Jogador", "Jogador_y": "Jogador2"})
df_parcerias
```

```python
salvamento_csv(df_individual, "desempenho_individual")
salvamento_csv(df_times, "desempenho_times")
salvamento_csv(df_info, "info_partidas")
salvamento_csv(df_sumula_completa, "sumula_completa")
salvamento_csv(df_parcerias, "parcerias")
```
